### countByKey and countByValue give different results for RDD, same for List Object parallelized

```python
#create rdd from file
lines = sc.textFile("C:/SparkCourse/book.txt")
```

```python
#function to parse line to words of lower case and space as delimiter
def splitWords(line):
    return line.lower().split()
```

```python
testRdd = sc.parallelize(['a','b','c','d','e','a','a','e','b'])
```

```python
#Action to convert to a hash map python dictionary
count1 = testRdd.countByKey()
print (count1)
```

```python
#Action to convert to a python dict
count2 = testRdd.countByValue()
print (count2)
```

### countByKey and countByValue give different results for RDD

```python
#split line int multiple word rowobject elements of RDD
wordsRdd3 = lines.flatMap(splitWords)
count3 = wordsRdd3.countByKey()
print (count3)
```

```python
#split line int multiple word rowobject elements of RDD
wordsRdd4 = lines.flatMap(splitWords)
count3 = wordsRdd3.countByValue()
print (count3)
```
